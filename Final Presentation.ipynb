```python
from geopy.geocoders import Nominatim
import folium

# Define the location (Beijing, CN for China)
address = "Beijing, CN"

# Create a geolocator object
geolocator = Nominatim(user_agent="CN_explorer")

# Retrieve the geographical coordinates for Beijing
location = geolocator.geocode(address)

# Extract latitude and longitude
latitude = location.latitude
longitude = location.longitude

# Print the coordinates for Beijing
print('The geographical coordinates of Beijing, China are {}, {}.'.format(latitude, longitude))

# Create a map of Beijing using latitude and longitude values
map_beijing = folium.Map(location=[latitude, longitude], zoom_start=11)

# Add markers to the map for locations in Beijing (Replace with your Beijing data)
for lat, lng, county, name in zip(beijing_df['Latitude'], beijing_df['Longitude'], beijing_df['County'], beijing_df['Place Name']):
    label = '{}, {}'.format(county, name)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='red',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False
    ).add_to(map_beijing)

# Display the map of Beijing
map_beijing
```
